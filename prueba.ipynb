```python
import pyreadstat

# 1. Define la ruta de tu archivo
ruta_archivo = 'def.sav' 

# 2. Leemos solo los metadatos (títulos) para que sea instantáneo
_, metadatos = pyreadstat.read_sav(ruta_archivo, metadataonly=True)

# 3. Mostrar los resultados
print(f"El archivo tiene un total de {len(metadatos.column_names)} columnas.\n")
print("Lista de columnas:")

# Nota: Los archivos SPSS suelen tener un "nombre" de variable corto (ej. VAR01) 
# y una "etiqueta" descriptiva (ej. "Edad del cliente"). Este código te muestra ambos:
for nombre, etiqueta in zip(metadatos.column_names, metadatos.column_labels):
    if etiqueta:
        print(f"- {nombre} (Descripción: {etiqueta})")
    else:
        print(f"- {nombre}")
```

```python
import pandas as pd
import pyreadstat

# 1. Define la ruta de tu archivo (asegúrate de que sea la correcta)
ruta_archivo = 'def.sav'

# 2. Leemos SOLO la columna 'causa_muerte' para no saturar la memoria.
# apply_value_formats=True hace la magia de cambiar los códigos numéricos por su texto real.
df_causa, _ = pyreadstat.read_sav(
    ruta_archivo, 
    usecols=['Caudef'], 
    apply_value_formats=True
)

# 3. Hacemos un resumen de las causas y cuántas veces aparece cada una
conteo_causas = df_causa['Caudef'].value_counts(dropna=False)

# 4. Mostramos el resultado
print("Categorías encontradas en 'Caudef ' y cantidad de casos:\n")
print(conteo_causas)
```
